### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
first_document = establishments.find_one(query)
print("First document with a hygiene score of 20:")
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
results = db.establishments.find(query)
df = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame
# Print the number of rows in the DataFrame and display the first 10 rows
print(f"Number of rows in the DataFrame: {len(df)}")
print("First 10 rows of the DataFrame:")
print(df.head(10))

```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName": {"$regex": "London"}, "RatingValue": {"$gte": 4}}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of establishments in London with RatingValue >= 4: {count}")

# Display the first document in the results using pprint
result = establishments.find(query)
first_result = result.next() if count > 0 else None
if first_result:
    print("First document:")
    pprint(first_result)
else:
    print("No documents found.")
```

```python
# Convert the result to a Pandas DataFrame
results = db.establishments.find(query)
df = pd.DataFrame(list(results))

# Display the number of rows in the DataFrame and display the first 10
print(f"Number of rows in the DataFrame: {len(df)}")
print("First 10 rows of the DataFrame:")
print(df.head(10))

```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Look up "Penang Flavours" to get its latitude and longitude
penang_flavours = establishments.find_one({"BusinessName": "Penang Flavours"})
latitude = penang_flavours['geocode']['latitude']
longitude = penang_flavours['geocode']['longitude']

    # Define the lat/lng search parameters
degree_search = 0.01

    # Query to find nearby establishments
query = {
        "RatingValue": 5,
        "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
        "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search}
    }

    # Sort by hygiene score (ascending)
sort = [("scores.Hygiene", 1)]  # 1 for ascending order

    # Call only top 5 results.
limit = 5

    # Query and sort the results
results = establishments.find(query).sort(sort).limit(limit)

    # Print the results
for result in results:
        pprint(result)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
pipeline = [
    {
        '$match': {
            'scores.Hygiene': 0,
        }
    },
    # Group by Local Authority Name, Count the number of establishments in each group, Push the full document into an array (optional).
    {
        '$group': {
            '_id': '$LocalAuthorityName', 
            'count': {'$sum': 1},   
        }
    },
    # Sort by the count in descending order
    {
        '$sort': {'count': -1}  
    }
]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
num_documents = len(results)
print(f"Number of Local Authorities with matching establishments: {num_documents}")

# Print the first 10 results
print("First 10 rows of the DataFrame:")
print(df.head(10))
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
print("Converted the results to a DataFrame.")

# Display the number of rows in the DataFrame
num_rows = df.shape[0]
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df.head(10))
```
